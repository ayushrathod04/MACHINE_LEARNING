```python
import pandas as pd
import numpy as np
import sklearn.datasets
from sklearn.preprocessing import StandardScaler
import seaborn as sb
from sklearn import svm
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
loan_df= pd.read_csv('/content/loan_dataset.csv')
```

```python
loan_df.head()
```

```python
loan_df.shape
```

```python
loan_df.isnull().sum()
```

```python
loan_df=loan_df.dropna()
```

```python
loan_df['Loan_Status'].value_counts()
```

```python
loan_df.describe()
```

```python
loan_df.replace({'Loan_Status':{'N':0,'Y':1}},inplace=True)
```

```python
loan_df['Dependents'].value_counts()
```

```python
from os import replace
```

```python
loan_df = loan_df.replace(to_replace='3+' ,value='4')
```

```python
loan_df['Dependents'].value_counts()
```

```python
loan_df.head()
```

```python
loan_df1 = loan_df.replace({'Married':{'Yes':1,'No':0},'Gender':{'Male':1,'Female':0},
                            'Self_Employed':{'Yes':1,'No':0},'Property_Area':{'Rural':0,'Semiurban':1,'Urban':2},
                            'Education':{'Graduate':1,'Not Graduate':0}})
```

```python
print(loan_df1.head())
```

```python
loanx = loan_df1.drop(columns=['Loan_ID','Loan_Status'],axis=1)
loany = loan_df1['Loan_Status']
```

```python
print(loanx)
print(loany)
```

```python
loanx_train,loanx_test,loany_train,loany_test = train_test_split(loanx,loany,test_size=0.1,stratify=loany,random_state=2)
```

```python
print(loanx.shape,loanx_train.shape,loanx_test.shape)
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(loanx_train,loany_train)
```

```python
loan_df1.head()
```

```python
loanx_train_pred = classifier.predict(loanx_train)
loanx_train_acc = accuracy_score(loany_train,loanx_train_pred)
print(loanx_train_acc)
```

```python
loanx_test_pred = classifier.predict(loanx_test)
loanx_test_acc = accuracy_score(loany_test,loanx_test_pred)
print(loanx_test_acc)
```
