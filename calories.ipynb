```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as seabornInstance
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
from sklearn.metrics import r2_score
```

```python
calories= pd.read_csv('/content/calories.csv')
```

```python
calories.head()
```

```python
calories.shape
```

```python
calories.info()
```

```python
calories.isnull().sum()
```

```python
exercise= pd.read_csv('/content/exercise.csv')
```

```python
exercise.head()
```

```python
cal_data= pd.concat([exercise,calories['Calories']],axis=1)
```

```python
cal_data.head()
```

```python
cal_data.shape
```

```python
cal_data.info()
```

```python
cal_data.replace({'Gender':{'male':0,'female':1}},inplace=True)
```

```python
cal_data.head()
```

```python
x=cal_data.drop(['User_ID','Calories'],axis=1)
y=cal_data['Calories']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
cal_model=XGBRegressor()
```

```python
cal_model.fit(x_train,y_train)
```

```python
x_train_prediction = cal_model.predict(x_train)
x_train_acc = metrics.r2_score(y_train, x_train_prediction)  # Use y_train for true values
print("R squared error : ", x_train_acc)
```

```python
x_test_prediction=cal_model.predict(x_test)
```

```python
cal_input = (1.0,36.0,151.0,50.0,23.0,96.0,40.7)
cal_input_as_numpy_array = np.asarray(cal_input)
cal_input_reshaped = cal_input_as_numpy_array.reshape(1,-1)
cal_pred = cal_model.predict(cal_input_reshaped)
print(cal_pred)
```
