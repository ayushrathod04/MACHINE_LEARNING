```python
df1= pd.read_csv('/content/drive/MyDrive/Evansh services files/Train - Train (1).csv')
```

```python
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1['Outlet_Size'].value_counts()
```

```python
df1['Item_Weight'].fillna(df1['Item_Weight'].mean(),inplace=True)
```

```python
df1['Outlet_Size']=df1['Outlet_Size'].fillna('Medium')
```

```python
df1.isnull().sum()
```

```python
df1.head()
```

```python
df1=df1.replace({'Outlet_Size':{'Small':0,'Medium':1,'High':2}})
```

```python
df1=df1.replace({'Item_Fat_Content':{'low fat':'Low Fat','LF':'Low Fat','reg':'Regular'}})
```

```python
label=LabelEncoder()
```

```python
df1['Item_Identifier']=label.fit_transform(df1['Item_Identifier'])
df1['Item_Fat_Content']=label.fit_transform(df1['Item_Fat_Content'])
df1['Item_Type']=label.fit_transform(df1['Item_Type'])
df1['Outlet_Identifier']=label.fit_transform(df1['Outlet_Identifier'])
df1['Outlet_Size']=label.fit_transform(df1['Outlet_Size'])
df1['Outlet_Location_Type']=label.fit_transform(df1['Outlet_Location_Type'])
df1['Outlet_Type']=label.fit_transform(df1['Outlet_Type'])
```

```python
df1.head()
```

```python
df1.isnull().sum()
```

```python
x=df1.drop(columns=['Item_Outlet_Sales'],axis=1)
y=df1['Item_Outlet_Sales']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
model=XGBRegressor()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_pred=model.predict(x_train)
print(x_train_pred)
```

```python
score_1=metrics.r2_score(y_train,x_train_pred)
score_2=metrics.mean_absolute_error(y_train,x_train_pred)
print(score_1)
print(score_2)
```
