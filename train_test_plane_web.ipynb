```python
train_df = train_df.replace({'Gender':{'Female':0,'Male':1}})
train_df = train_df.replace({'Gender':{'Female':0,'Male':1}})
train_df = train_df.replace({'Customer Type':{'Loyal Customer':0,'disloyal Customer':1}})
train_df = train_df.replace({'Type of Travel':{'Personal Travel':0,'Business travel':1}})
train_df = train_df.replace({'Class':{'Business':0,'Eco':1,'Eco Plus':2}})
train_df = train_df.replace({'satisfaction':{'neutral or dissatisfied':0,'satisfied':1}})
```

```python
train_df.head()
```

```python
test_df = test_df.drop(columns = ['id','Unnamed: 0'], axis=1)
```

```python
test_df.head()
```

```python
x = train_df.drop(columns = ['satisfaction','id','Unnamed: 0'], axis=1)
y = train_df['satisfaction']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.predict(x_train)
accuracy_score(y_train, model.predict(x_train))
```
