```python
data=pd.read_csv('/content/StudentsPerformance - StudentsPerformance.csv')
```

```python
data.head()
```

```python
sns.set_theme()
```

```python
# Create subplots for each gender
genders = data['gender'].unique()
fig, axes = plt.subplots(1, len(genders), figsize=(16, 6), sharex=True, sharey=True)

for ax, gender in zip(axes, genders):
    subset = data[data['gender'] == gender]
    sns.scatterplot(
        data=subset,
        x="math score", y="reading score",
        hue="lunch", style="test preparation course", size="writing score",
        sizes=(40, 200), alpha=0.7, palette="muted",
        ax=ax
    )
    ax.set_title(f"Gender: {gender}")

# Set overall plot title and show the plot
plt.suptitle("Student Performance: Math vs. Reading Scores by Gender")
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Math score box plot
sns.boxplot(x="test preparation course", y="math score", data=data, ax=axes[0])
axes[0].set_title("Math Scores by Test Preparation Course")

# Reading score box plot
sns.boxplot(x="test preparation course", y="reading score", data=data, ax=axes[1])
axes[1].set_title("Reading Scores by Test Preparation Course")

# Writing score box plot
sns.boxplot(x="test preparation course", y="writing score", data=data, ax=axes[2])
axes[2].set_title("Writing Scores by Test Preparation Course")

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
sns.lmplot(
    data=data,
    x="math score", y="reading score",
    hue="test preparation course",
    col="test preparation course",
    height=6,
    aspect=1
)

# Set overall plot title
plt.suptitle("Reading Scores vs. Math Scores by Test Preparation Course", y=1.02)
plt.show()
```

```python
sns.histplot(
    data=data,
    x="reading score",
    hue="parental level of education",
    multiple="stack",
    palette="muted"
)

# Set plot title and labels
plt.title("Distribution of Reading Scores by Parental Level of Education")
plt.xlabel("Reading Score")
plt.ylabel("Frequency")
plt.legend(title="Parental Level of Education", bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.histplot(
    data=data,
    x="reading score",
    hue="parental level of education",
    multiple="stack",
    palette="muted"
)

# Set plot title and labels
plt.title("Distribution of Reading Scores by Parental Level of Education")
plt.xlabel("Reading Score")
plt.ylabel("Frequency")
plt.legend(title="Parental Level of Education", bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
g = sns.FacetGrid(data, col="race/ethnicity", col_wrap=3, height=5)
g.map(sns.histplot, "reading score", bins=10, kde=False, color="blue")

# Set overall plot title
plt.subplots_adjust(top=0.9)
g.fig.suptitle("Distribution of Reading Scores by Race/Ethnicity")

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
```

```python
amazon_df = pd.read_csv('/content/amazon.csv')

# Clean and convert relevant columns to numeric types
amazon_df['discounted_price'] = amazon_df['discounted_price'].str.replace('â‚¹', '').str.replace(',', '').astype(float)
amazon_df['actual_price'] = amazon_df['actual_price'].str.replace('â‚¹', '').str.replace(',', '').astype(float)
amazon_df['discount_percentage'] = amazon_df['discount_percentage'].str.replace('%', '').astype(float)
amazon_df['rating'] = amazon_df['rating'].replace('|', np.nan).astype(float)
amazon_df['rating_count'] = amazon_df['rating_count'].str.replace(',', '').astype(float)
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(14, 6))
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# Discounted Price
axes[0, 0].hist(amazon_df['discounted_price'].dropna(), bins=30, color='skyblue', edgecolor='black')
axes[0, 0].set_title('Distribution of Discounted Prices')
axes[0, 0].set_xlabel('Discounted Price')
axes[0, 0].set_ylabel('Frequency')

# Actual Price
axes[0, 1].hist(amazon_df['actual_price'].dropna(), bins=30, color='lightgreen', edgecolor='black')
axes[0, 1].set_title('Distribution of Actual Prices')
axes[0, 1].set_xlabel('Actual Price')
axes[0, 1].set_ylabel('Frequency')

# Discount Percentage
axes[0, 2].hist(amazon_df['discount_percentage'].dropna(), bins=30, color='salmon', edgecolor='black')
axes[0, 2].set_title('Distribution of Discount Percentages')
axes[0, 2].set_xlabel('Discount Percentage')
axes[0, 2].set_ylabel('Frequency')

# Rating
axes[1, 0].hist(amazon_df['rating'].dropna(), bins=30, color='orchid', edgecolor='black')
axes[1, 0].set_title('Distribution of Ratings')
axes[1, 0].set_xlabel('Rating')
axes[1, 0].set_ylabel('Frequency')

# Rating Count
axes[1, 1].hist(amazon_df['rating_count'].dropna(), bins=30, color='gold', edgecolor='black')
axes[1, 1].set_title('Distribution of Rating Counts')
axes[1, 1].set_xlabel('Rating Count')
axes[1, 1].set_ylabel('Frequency')

# Hide the last subplot (1, 2)
axes[1, 2].axis('off')

plt.tight_layout()
plt.show()
```

```python
# Boxplots for numerical columns
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Discounted Price
sns.boxplot(y=amazon_df['discounted_price'], ax=axes[0], color='skyblue')
axes[0].set_title('Boxplot of Discounted Prices')

# Actual Price
sns.boxplot(y=amazon_df['actual_price'], ax=axes[1], color='lightgreen')
axes[1].set_title('Boxplot of Actual Prices')

plt.tight_layout()
plt.show()
```

```python
# Countplot for categories
plt.figure(figsize=(14, 6))
sns.countplot(y=amazon_df['category'], order=amazon_df['category'].value_counts().index)
plt.title('Count of Products per Category')
plt.show()
```

```python
# Scatter plots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Discounted Price vs Actual Price
sns.scatterplot(x=amazon_df['discounted_price'], y=amazon_df['actual_price'], ax=axes[0])
axes[0].set_title('Discounted Price vs Actual Price')

# Discount Percentage vs Rating
sns.scatterplot(x=amazon_df['discount_percentage'], y=amazon_df['rating'], ax=axes[1])
axes[1].set_title('Discount Percentage vs Rating')

plt.tight_layout()
plt.show()
```

```python
# Bar plot for average rating per category
plt.figure(figsize=(14, 6))
avg_rating_per_category = amazon_df.groupby('category')['rating'].mean().sort_values(ascending=False)
sns.barplot(x=avg_rating_per_category, y=avg_rating_per_category.index, palette='viridis')
plt.title('Average Rating per Category')
plt.xlabel('Average Rating')
plt.ylabel('Category')
plt.show()
```
